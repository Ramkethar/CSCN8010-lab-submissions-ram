  four different types of data visualizations using Matplotlib and Seaborn :


```python
import matplotlib.pyplot as mpb

# Sample data
gender_labels = ['Male', 'Female']
gender_counts = [45, 55]  # These numbers represent the counts or percentages of males and females.

# Create a pie chart
mpb.figure(figsize=(6, 6))
mpb.pie(gender_counts, labels=gender_labels, autopct='%1.1f%%', startangle=140)
mpb.title('Male vs. Female Ratio')

# Display the pie chart
mpb.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

mpb.show()
```

gender_labels = the labels for the two categories, "Male" and "Female."

gender_counts = the corresponding counts or percentages. 

```python
import seaborn as sbn
import matplotlib.pyplot as mpb  # Changed plt to mpb

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [1000, 1200, 1300, 1100, 1400, 1500]

# Create a line plot using Seaborn
sbn.set(style="whitegrid")
mpb.figure(figsize=(8, 6))
sbn.lineplot(x=months, y=sales, marker='o', color='b')
mpb.xlabel('Month')
mpb.ylabel('Sales')
mpb.title('Monthly Sales Trend')

mpb.show()
```

months = the months of the year.
sales = the corresponding sales figures for each month.

```python
import seaborn as sbn
import matplotlib.pyplot as mpb  # Changed plt to mpb
import pandas as pd

# Sample data (dates and daily new cases)
dates = pd.date_range(start='2023-01-01', periods=30, freq='D')
daily_new_cases = [10, 12, 15, 18, 25, 28, 35, 42, 55, 60, 70, 75, 80, 82, 90, 85, 78, 70, 60, 55, 50, 45, 42, 38, 30, 25, 22, 18, 15, 12]

# Create a line plot using Seaborn
sbn.set(style="whitegrid")
mpb.figure(figsize=(12, 6))
sbn.lineplot(x=dates, y=daily_new_cases, marker='o', color='r', linestyle='-')
mpb.xlabel('Date')
mpb.ylabel('Daily New Cases')
mpb.title('Daily New COVID-19 Cases Trend (Example)')

# Rotate x-axis labels for better readability
mpb.xticks(rotation=45)

mpb.show()
```

dates = a date range for the 30 days in our example.
daily_new_cases = the corresponding daily new COVID-19 cases for each day. 

```python
import matplotlib.pyplot as mpb

# Sample data
countries = ['Canada', 'India']
population = [37590000, 1393409038]  # Population figures are in millions

# Create a bar chart
mpb.figure(figsize=(8, 6))
mpb.bar(countries, population, color=['blue', 'orange'])
mpb.xlabel('Country')
mpb.ylabel('Population (Millions)')
mpb.title('Population Comparison: Canada vs. India')

mpb.show()
```

countries  = the names of the two countries, Canada and India.
population = the corresponding population sizes in millions.
